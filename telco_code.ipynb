```python
df = df.withColumn("sms_sent", col("sms_sent").cast("int")) 
df = df.withColumn("plan_monthly_price", col("plan_monthly_price").cast("float"))
```

```python
df.withColumn("region",
    when(col("country").isin("Nigeria", "Kenya", "South Africa"), "Africa")
    .when(col("country").isin("UK", "Germany", "France"), "Europe")
    .when(col("country").isin("USA", "Canada"), "North America")
    .otherwise("Other")
)
```

```python
for column in df.columns:
    print(column, 'Nulls:', df.filter(df[column].isNull()).count())
```

```python
df = df.fillna({
                "plan_name": 'Unknown',
                "plan_monthly_price": 0,
                "data_used_mb": 0,
                "voice_minutes_used": 0.0,
                "sms_sent": 0


})
```

```python
df = df.na.drop(subset = ["first_name"])
```

```python
for col in df.columns:
    print(col, 'Nulls:', df.filter(df[col].isNull()).count())
```

```python
df.count()
```

```python
df = df.dropDuplicates(subset= ["customer_id"])
```

```python
df.count()
```

```python
df.show(5)
```

```python
#check for min and max subscription dates
df.agg({"subscription_start_date": "min", "subscription_end_date": "max"}).show()
```

```python
from pyspark.sql.functions import months_between, round,col
df = df.withColumn(
    "subscription_months",
    round(months_between(col("subscription_end_date"), col("subscription_start_date")))
)


df = df.withColumn(
    "revenue",
    col("plan_monthly_price") * col("subscription_months")
)
```

```python
df =df.withColumn("customer_category",
                  when(col("revenue") < 100, "Low")
    .when((col("revenue") >= 100) & (col("revenue") < 300), "Medium")
    .otherwise("High")
)
```

```python
#region column
df =df.withColumn("region",
                    when(col("country").isin("Nigeria", "Kenya", "South Africa"), "Africa")
        .when(col("country").isin("UK", "Germany", "France"), "Europe")
        .when(col("country").isin("USA", "Canada"), "North America")
        .otherwise("Other")
    )
                  

```

```python
df.show(5)
```

```python
customer_dim = df.select(
    "customer_id",
    "first_name",
    "last_name",
    "country",
    "region",
    "signup_date",
    "status",
    "customer_category"
)
customer_dim = customer_dim.dropDuplicates(subset= ["customer_id"])
```

```python
subscription_dim = df.select(
    "customer_id",
    "subscription_id",
    "subscription_start_date",
    "subscription_end_date",
    "subscription_months"
).dropDuplicates(subset=["subscription_id"])
subscription_dim = subscription_dim.dropDuplicates(subset= ["subscription_id"])
```

```python
plan_dim = df.select(
    "plan_id",
    "subscription_id",
    "plan_name",
    "plan_type",
    "plan_monthly_price",
    "plan_data_limit_gb",
    "plan_voice_minutes",
    "network_type"
).dropDuplicates(subset=["plan_id"])
plan_dim = plan_dim.dropDuplicates(subset= ["plan_id"])
```

```python
usage_fact = df.select(
    "customer_id",
    "subscription_id",
    "plan_id",
    "usage_date",
    "data_used_mb",
    "voice_minutes_used",
    "sms_sent",
    "revenue",
  
)
usage_fact = usage_fact.withColumn("usage_id", monotonically_increasing_id())
usage_fact = usage_fact.select(
    "usage_id",
    "customer_id",
    "subscription_id",
    "plan_id",
    "usage_date",
    "data_used_mb",
    "voice_minutes_used",
    "sms_sent",
    "revenue"
)
```

```python
from pyspark.sql.functions import col, dayofmonth, weekofyear, month, quarter, year, sequence, explode, to_date, lit
date_dim = spark.sql("""
    SELECT explode(sequence(to_date('2025-01-01'), to_date('2027-01-01'), interval 1 day)) as date
""")

# Add day, week, month, quarter, year columns
date_dim = date_dim.withColumn("day", dayofmonth(col("date"))) \
                   .withColumn("week", weekofyear(col("date"))) \
                   .withColumn("month", month(col("date"))) \
                   .withColumn("quarter", quarter(col("date"))) \
                   .withColumn("year", year(col("date")))
```

```python
def get_db_connection():
    connection = psycopg2.connect(host ='localhost',
                                  password ='postgres',
                                  database = 'telcosphere',
                                  user = 'postgres')
    return connection
```

```python
def create_table():
    conn = get_db_connection()
    cursor = conn.cursor()
    table_query = '''   CREATE SCHEMA stg;
                        CREATE SCHEMA edw;

                        CREATE TABLE stg.customer_dim(
                            customer_id varchar (50) PRIMARY KEY,
                            first_name varchar (50),
                            last_name varchar (50),
                            country varchar (50),
                            region varchar (50),
                            signup_date date,
                            status varchar (50),
                            customer_category varchar (50)
                    );

                        CREATE TABLE stg.subscription_dim(
                            customer_id varchar (50),
                            subscription_id varchar (50) PRIMARY KEY,
                            subscription_start_date date,
                            subscription_end_date date,
                            subscription_months int,
                            FOREIGN KEY (customer_id) REFERENCES stg.customer_dim(customer_id)
                            );


                        CREATE TABLE stg.plan_dim(
                            plan_id varchar (50) PRIMARY KEY,
                            subscription_id varchar (50),
                            plan_name varchar (50),
                            plan_type varchar (50),
                            plan_monthly_price FLOAT,
                            plan_data_limit_gb int,
                            plan_voice_minutes int,
                            network_type varchar (50),
                            FOREIGN KEY (subscription_id) REFERENCES stg.subscription_dim(subscription_id)
                        );  


                        CREATE TABLE stg.usage_fact(
                            usage_id varchar(50) PRIMARY KEY,
                            customer_id varchar(50),
                            subscription_id varchar(50),
                            plan_id varchar(50),
                            usage_date date,
                            data_used_mb FLOAT,
                            voice_minutes_used FLOAT,
                            sms_sent int,
                            revenue FLOAT,
                            FOREIGN KEY (customer_id) REFERENCES stg.customer_dim(customer_id),
                            FOREIGN KEY (subscription_id) REFERENCES stg.subscription_dim(subscription_id),
                            FOREIGN KEY (plan_id) REFERENCES stg.plan_dim(plan_id) 
                        );



                          CREATE TABLE edw.date_dim(
                            date date PRIMARY KEY,
                            day INT,
                            week INT,
                            month INT,
                            quarter INT,
                            year INT

                        );



                        CREATE TABLE edw.customer_dim(
                            customer_id varchar (50) PRIMARY KEY,
                            first_name varchar (50),
                            last_name varchar (50),
                            country varchar (50),
                            region varchar (50),
                            signup_date date,
                            status varchar (50),
                            customer_category varchar (50),
                            FOREIGN KEY (signup_date) REFERENCES edw.date_dim(date)

                        );

                        CREATE TABLE edw.subscription_dim(
                            customer_id varchar (50),
                            subscription_id varchar (50) PRIMARY KEY,
                            subscription_start_date date,
                            subscription_end_date date,
                            subscription_months int,
                            FOREIGN KEY (customer_id) REFERENCES edw.customer_dim(customer_id),
                            FOREIGN KEY (subscription_start_date) REFERENCES edw.date_dim(date),
                            FOREIGN KEY (subscription_end_date) REFERENCES edw.date_dim(date)
                        );


                        CREATE TABLE edw.plan_dim(
                            plan_id varchar (50) PRIMARY KEY,
                            subscription_id varchar (50),
                            plan_name varchar (50),
                            plan_type varchar (50),
                            plan_monthly_price FLOAT,
                            plan_data_limit_gb int,
                            plan_voice_minutes int,
                            network_type varchar (50),
                            FOREIGN KEY (subscription_id) REFERENCES edw.subscription_dim(subscription_id)
                        );  


                        CREATE TABLE edw.usage_fact(
                            usage_id varchar(50) PRIMARY KEY,
                            customer_id varchar(50),
                            subscription_id varchar(50),
                            plan_id varchar(50),
                            usage_date date,
                            data_used_mb FLOAT,
                            voice_minutes_used FLOAT,
                            sms_sent int,
                            revenue FLOAT,
                            FOREIGN KEY (customer_id) REFERENCES edw.customer_dim(customer_id),
                            FOREIGN KEY (subscription_id) REFERENCES edw.subscription_dim(subscription_id),
                            FOREIGN KEY (plan_id) REFERENCES edw.plan_dim(plan_id),
                            FOREIGN KEY (usage_date) REFERENCES edw.date_dim(date)
                        );

'''
    
    cursor.execute(table_query)
    conn.commit()
    cursor.close()
    conn.close()
```

```python
create_table()
```

```python
##loading into db
url = "jdbc:postgresql://localhost:5432/telcosphere"
user = "postgres"
password = "postgres"

def load_to_postgres(df, table_name):
    df.write.format("jdbc")\
    .option("url", url)\
    .option("dbtable", table_name)\
    .option("user", user)\
    .option("password", password)\
    .option("driver", "org.postgresql.Driver")\
    .mode("append")\
    .save()
```

```python
load_to_postgres(customer_dim, "stg.customer_dim")
load_to_postgres(subscription_dim, "stg.subscription_dim")
load_to_postgres(plan_dim, "stg.plan_dim")
load_to_postgres(usage_fact, "stg.usage_fact")
load_to_postgres(date_dim, "edw.date_dim")
load_to_postgres(customer_dim, "edw.customer_dim")
load_to_postgres(subscription_dim, "edw.subscription_dim")
load_to_postgres(plan_dim, "edw.plan_dim")
load_to_postgres(usage_fact, "edw.usage_fact")
```
